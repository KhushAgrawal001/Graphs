```python
import matplotlib.pyplot as plt
import numpy as np
def plot_trig_functions(x_min, x_max, num_points):
  x = np.linspace(x_min, x_max, num_points)

  # subplots
  fig, axes = plt.subplots(2, 3, figsize=(12, 6))
  axes = axes.ravel()
  # sine function
  axes[0].plot(x, np.sin(x), label="Sine")
  axes[0].set_title("Sine")
  # cosine function
  axes[1].plot(x, np.cos(x), label="Cosine")
  axes[1].set_title("Cosine")
  safe_mask = np.abs(np.cos(x)) > 1e-6  
  axes[2].plot(x[safe_mask], np.tan(x[safe_mask]), label="Tangent")
  axes[2].set_title("Tangent")
  axes[2].set_ylim(-10, 10)
  # arcsine function
  axes[3].plot(x, np.arcsin(x), label="Arcsine")
  axes[3].set_title("Arcsine")
  axes[3].set_ylim(-np.pi/2, np.pi/2)  
  # arccosine function
  axes[4].plot(x, np.arccos(x), label="Arccosine")
  axes[4].set_title("Arccosine")
  axes[4].set_ylim(-np.pi, np.pi/2) 
  # arctangent function
  axes[5].plot(x, np.arctan(x), label="Arctangent")
  axes[5].set_title("Arctangent")
  # Add labels and show the plot
  for ax in axes:
    ax.legend()
    ax.grid(True)
  plt.xlabel("X")
  plt.tight_layout()
  plt.show()

# Set plot parameters
x_min = -5 * np.pi
x_max = 5 * np.pi
num_points = 400

plot_trig_functions(x_min, x_max, num_points)
```

```python
plt.style.use('dark_background')
def plot_trig_function(x_min, x_max, num_points, func_name, func):
  x = np.linspace(x_min, x_max, num_points)
  plt.plot(x, func(x), label=func_name)
  plt.xlabel("X")
  plt.ylabel(func_name)
  plt.title(func_name + " Function")
  plt.grid(True)
  plt.legend()
  plt.show()

# Define functions for sine, cosine, and tangent

def sine(x):
  return np.sin(x)
def cosine(x):
  return np.cos(x)
def tangent(x):
  return np.tan(x)

# plot parameters

x_min = -2 * np.pi
x_max = 2 * np.pi
num_points = 200
plot_trig_function(x_min, x_max, num_points, "Sine", sine)
plot_trig_function(x_min, x_max, num_points, "Cosine", cosine)
plot_trig_function(x_min, x_max, num_points, "Tangent", tangent)
```
